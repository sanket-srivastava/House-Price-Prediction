```python
df = df[['OverallQual','YearBuilt','TotalBsmtSF','1stFlrSF','GrLivArea','FullBath','TotRmsAbvGrd','GarageCars','GarageArea','SalePrice']]
```

```python
df.head()
```

```python
sns.heatmap(data=df.corr(),vmin=-1,vmax=1,annot=True)
```

```python
sns.boxplot(x='GarageCars',y='GarageArea',data=df)
```

```python
df.drop(['TotalBsmtSF','FullBath','TotRmsAbvGrd','GarageCars'],axis=1,inplace=True)
```

```python
df.dropna(inplace=True)
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(df.drop('SalePrice',axis=1),df['SalePrice'],test_size=0.3)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(xtrain,ytrain)
```

```python
p = model.predict(xtest)
```

```python
from sklearn.metrics import mean_squared_error
error = np.sqrt(mean_squared_error(ytest,p))
```
